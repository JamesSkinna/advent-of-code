```python
def check_access(grid: list[str], pos: list):
    roll_count = 0

    for row in range(pos[0] - 1, pos[0] + 2):
        for col in range(pos[1] - 1, pos[1] + 2):
            if not (row == pos[0] and col == pos[1]):
                if row > -1 and col > -1:
                    try:
                        if grid[row][col] == "@":
                            roll_count += 1
                            if roll_count >= 4:
                                return False
                    except IndexError:
                        pass

    return True
                

test_grid = test_input.split("\n")
print(check_access(test_grid, [0, 7]))
print(check_access(test_grid, [0, 2]))
    
```

```python
def count_accessible_p1(input: str):
    accessible_count = 0
    grid = input.split("\n")
    for row in range(len(grid)):
        for col in range(len(grid[0])):
            if grid[row][col] == "@":
                if check_access(grid, [row, col]):
                    accessible_count += 1

    return accessible_count

count_accessible_p1(test_input)
```

```python
def count_accessible_p2(input: str):
    has_changed = True
    accessible_count = 0
    grid = input.split("\n")

    while has_changed:
        has_changed = False

        for row in range(len(grid)):
            for col in range(len(grid[0])):
                if grid[row][col] == "@":
                    if check_access(grid, [row, col]):
                        accessible_count += 1
                        grid[row] = grid[row][:col] + "." + grid[row][col+1:]    # clear the roll
                        has_changed = True

    return accessible_count

count_accessible_p2(test_input)
```
