# Part 1

```python
def is_valid_id_p1(id: str):
    # Odd lengths are always valid
    if len(id) % 2 != 0:
        return True
    # split into 2 parts
    half_pointer = len(id) // 2
    p1 = id[0:half_pointer]
    p2 = id[half_pointer:]
    if p1 == p2:
        return False
    return True

print(is_valid_id_p1("12521"))
print(is_valid_id_p1("1010"))
print(is_valid_id_p1("1188511885"))
```

```python
def sum_invalids_p1(input: str):
    ranges = input.split(",")

    invalid_sum = 0

    for r in ranges:
        start, end = r.split("-")

        for id in range(int(start), int(end) + 1, 1):
            if not is_valid_id_p1(str(id)):
                invalid_sum += id

    return invalid_sum

sum_invalids_p1(test_input)
```

# Part 2

```python
# Efficient way to find factors
ranges = input.split(",")
max_len = 0

# Find longest length
for r in ranges:
    start, end = r.split("-")
    if len(end) > max_len:
        max_len = len(end)

# Calculate all factors (ignoring 1)
factors_lookup = {1: []}

for i in range(2, max_len + 1):
    for j in range(2, i + 1):
        if i % j == 0:
            if i in factors_lookup:
                factors_lookup[i].append(j)
            else:
                factors_lookup[i] = [j]

print(factors_lookup)
```

```python
def is_valid_id_p2(id: str):
    length = len(id)
    factors = factors_lookup[length]

    for factor in factors:
        part_len = int(length / factor)
        parts = ([id[i:i + part_len] for i in range(0, len(id), part_len)])
        unique_parts = set(parts)
        if len(unique_parts) == 1:
            return False
        
    return True


print(is_valid_id_p2("12521"))
print(is_valid_id_p2("1010"))
print(is_valid_id_p2("1188511885"))
print(is_valid_id_p2("824824824"))
print(is_valid_id_p2("999"))
print(is_valid_id_p2("9991"))
```

```python
def sum_invalids_p2(input: str):
    ranges = input.split(",")

    invalid_sum = 0

    for r in ranges:
        start, end = r.split("-")

        for id in range(int(start), int(end) + 1, 1):
            if not is_valid_id_p2(str(id)):
                invalid_sum += id

    return invalid_sum

sum_invalids_p2(test_input)
```
