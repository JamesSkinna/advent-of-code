```python
from tqdm import tqdm

def make_fresh_dict(fresh_ranges: list[str]):
    fresh_dict = {}

    for r in tqdm(fresh_ranges):
        lower, higher = r.split("-")

        for id in range(int(lower), int(higher) + 1):
            if id not in fresh_dict:
                fresh_dict[id] = True

    return fresh_dict


def count_fresh_p1(input: str):

    fresh_ranges_str, ids_str = input.split("\n\n")
    fresh_ranges = fresh_ranges_str.split("\n")
    ids = ids_str.split("\n")

    fresh_dict = make_fresh_dict(fresh_ranges)

    fresh_count = 0

    for id in tqdm(ids):
        if int(id) in fresh_dict:
            fresh_count += 1

    return fresh_count
```
