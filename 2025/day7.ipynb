# Part 1

```python
def count_splits_p1(input: str):
    rows = input.split("\n")
    prev_beam_positions = set([rows[0].index("S")])
    split_count = 0
    new_beam_positions = []

    for row in rows[1:]:
        for beam_pos in prev_beam_positions:
            if row[beam_pos] == "^":
                split_count += 1
                new_beam_positions.append(beam_pos + 1)
                new_beam_positions.append(beam_pos - 1)
            else:
                new_beam_positions.append(beam_pos)
        prev_beam_positions = set(new_beam_positions)
        new_beam_positions = []

    return split_count

count_splits_p1(test_input)
    
```

# Part 2 - V1, Recursive

```python

def traverse_timeline(rows: list[str], prev_beam_position: int):
    result = 0
    if rows:
        if rows[0][prev_beam_position] == "^":
            result += traverse_timeline(rows[1:], prev_beam_position + 1)
            result += traverse_timeline(rows[1:], prev_beam_position - 1) 
        else:
            result += traverse_timeline(rows[1:], prev_beam_position) 
        return result
    return 1


def count_splits_p2(input: str):
    rows = input.split("\n")
    prev_beam_position = rows[0].index("S")
    return traverse_timeline(rows[1:], prev_beam_position)

count_splits_p2(test_input)
    
```

# Part 2 - V2, Non-Recursive

```python
from tqdm import tqdm

def count_splits_p2_v2(input: str):
    rows = input.split("\n")
    start_index = rows[0].index("S")
    prev_beam_positions = {start_index: 1}   # only 1 beam to start
    new_beam_positions = {}

    for row in tqdm(rows[1:]):
        for beam_pos, vals in prev_beam_positions.items():
            if row[beam_pos] == "^":
                if beam_pos + 1 in new_beam_positions:
                    new_beam_positions[beam_pos + 1] += vals
                else:
                    new_beam_positions[beam_pos + 1] = vals
                if beam_pos - 1 in new_beam_positions:
                    new_beam_positions[beam_pos - 1] += vals
                else:
                    new_beam_positions[beam_pos - 1] = vals
            else:
                if beam_pos in new_beam_positions:
                    new_beam_positions[beam_pos] += vals
                else:
                    new_beam_positions[beam_pos] = vals
        prev_beam_positions = new_beam_positions
        new_beam_positions = {}

    return sum([val for val in prev_beam_positions.values()])

count_splits_p2_v2(test_input)
    
```
