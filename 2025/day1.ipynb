# Part 1 - Point at Zero

```python
def count_zeros(input: str, dial_pos_start: int = 50, min: int = 0, max: int = 100, count_num: int = 0):
    rotations = input.split("\n")

    count = 0
    dial_pos = dial_pos_start

    for rotation in rotations:
        dir = rotation[0]
        dist = int(rotation[1:]) % max  # mod 100 incase multi rotatons past 99

        if dir == "L":
            dial_pos -= dist
        else:
            dial_pos += dist

        if dial_pos >= max:
            dial_pos -= max
        elif dial_pos < min:
            dial_pos += max

        if dial_pos == count_num:
            count += 1
        
    return count
```

# Part 2 - Move past Zero

```python
def count_past_zero(input: str, dial_pos_start: int = 50, min: int = 0, max: int = 100, count_num: int = 0):
    rotations = input.split("\n")

    count = 0
    dial_pos = dial_pos_start

    for rotation in rotations:
        dir = rotation[0]
        dist = int(rotation[1:]) % max  # mod 100 incase multi rotatons past 99
        num_extra_passes =  int(rotation[1:]) // max    # count extra passes past zero

        count += num_extra_passes

        prev_dial_pos = dial_pos

        if dir == "L":
            dial_pos -= dist
        else:
            dial_pos += dist

        if (
            (prev_dial_pos > count_num > dial_pos)
            or (prev_dial_pos < count_num < dial_pos)
            or (prev_dial_pos < max < dial_pos)
            or (prev_dial_pos > max > dial_pos)
        ):
            count += 1

        if dial_pos >= max:
            dial_pos -= max
        elif dial_pos < min:
            dial_pos += max

        if dial_pos == count_num:
            count += 1
        
    return count

count_past_zero(test_input)
```
